```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('IIM_database.csv')
```

```python
emotions = []
for x in df['filename']:
    emotions.append(x[-7:-4])

df['emotions'] = emotions
```

```python
df
```

```python
emotion_lst = df['emotions'].unique()

for emotion in emotion_lst:
    df[emotion] = [0 for i in range(df.shape[0])]
df['RESULT'] = ['NA' for i in range(df.shape[0])]
```

```python
df
```

```python
# mapping actions units with their respective index in pandas dataframe

facs_to_index = {
    'AU01_r' :0,
    'AU02_r' :1,
    'AU04_r' :2,
    'AU05_r' :3,
    'AU06_r' :4,
    'AU07_r' :5,
    'AU09_r' :6,
    'AU010_r' :7,
    'AU012_r' :8,
    'AU014_r' :9,
    'AU015_r' :10,
    'AU017_r' :11,
    'AU020_r' :12,
    'AU023_r' :13,
    'AU025_r' :14,
    'AU026_r' :15,
    'AU045_r' :16,
    'filename':17,
    'emotions':18,
    'NUT':19,
    'FER':20,
    'SUR':21,
    'ANG':22,
    'DIS':23,
    'SAD':24,
    'HPY':25,
    'RESULT':26
}

thresold = 0.8

result = []

for index in range(df.shape[0]):

    # We are using AU description as provided in radboud paper

    #This is for angry
    if df.iloc[index,facs_to_index['AU04_r']] >= thresold and df.iloc[index,facs_to_index['AU05_r']] >= thresold and df.iloc[index,facs_to_index['AU023_r']] >= thresold and df.iloc[index,facs_to_index['AU017_r']] >= thresold and df.iloc[index,facs_to_index['AU07_r']] >= thresold:
        df.iloc[index,facs_to_index['ANG']] = 1 
        if df.iloc[index,facs_to_index['emotions']] == 'ANG':
            df.iloc[index,facs_to_index['RESULT']] = 'T'
            
        else:
            df.iloc[index,facs_to_index['RESULT']] = 'F'

    # This is for Disgusted
    if df.iloc[index,facs_to_index['AU09_r']] >= thresold and df.iloc[index,facs_to_index['AU025_r']] >= thresold and df.iloc[index,facs_to_index['AU010_r']] >= thresold:
        df.iloc[index,facs_to_index['DIS']] = 1 
        if df.iloc[index,facs_to_index['emotions']] == 'DIS' and df.iloc[index,facs_to_index['RESULT']]=='NA' :
            df.iloc[index,facs_to_index['RESULT']] = 'T'
            
        else:
            df.iloc[index,facs_to_index['RESULT']] = 'F'

    # This is for fearful
    if df.iloc[index,facs_to_index['AU04_r']] >= thresold and df.iloc[index,facs_to_index['AU01_r']] >= thresold and df.iloc[index,facs_to_index['AU02_r']] >= thresold and df.iloc[index,facs_to_index['AU05_r']] >= thresold and df.iloc[index,facs_to_index['AU025_r']] >= thresold and df.iloc[index,facs_to_index['AU020_r']] >= thresold:
        df.iloc[index,facs_to_index['FER']] = 1
        if df.iloc[index,facs_to_index['emotions']] == 'FER' and df.iloc[index,facs_to_index['RESULT']]=='NA':
            df.iloc[index,facs_to_index['RESULT']] = 'T'
            
        else:
            df.iloc[index,facs_to_index['RESULT']] = 'F'

    # This is for Happy
    if df.iloc[index,facs_to_index['AU06_r']] >= thresold and df.iloc[index,facs_to_index['AU012_r']] >= thresold and df.iloc[index,facs_to_index['AU025_r']] >= thresold:
        df.iloc[index,facs_to_index['HPY']] = 1
        if df.iloc[index,facs_to_index['emotions']] == 'HPY' and df.iloc[index,facs_to_index['RESULT']]=='NA':
            df.iloc[index,facs_to_index['RESULT']] = 'T'
            
        else:
            df.iloc[index,facs_to_index['RESULT']] = 'F'

    # This is for sad
    if df.iloc[index,facs_to_index['AU04_r']] >= thresold and df.iloc[index,facs_to_index['AU01_r']] >= thresold and df.iloc[index,facs_to_index['AU015_r']] >= thresold and df.iloc[index,facs_to_index['AU017_r']] >= thresold:
        df.iloc[index,facs_to_index['SAD']] = 1
        if df.iloc[index,facs_to_index['emotions']] == 'SAD' and df.iloc[index,facs_to_index['RESULT']]=='NA':
            df.iloc[index,facs_to_index['RESULT']] = 'T'
            
        else:
            df.iloc[index,facs_to_index['RESULT']] = 'F'
    # This is for surprised
    if df.iloc[index,facs_to_index['AU01_r']] >= thresold and df.iloc[index,facs_to_index['AU05_r']] >= thresold and df.iloc[index,facs_to_index['AU02_r']] >= thresold and df.iloc[index,facs_to_index['AU026_r']] >= thresold:
        df.iloc[index,facs_to_index['SUR']] = 1
        if df.iloc[index,facs_to_index['emotions']] == 'SUR' and df.iloc[index,facs_to_index['RESULT']]=='NA':
            df.iloc[index,facs_to_index['RESULT']] = 'T'
        else:
            df.iloc[index,facs_to_index['RESULT']] = 'F'

        
```
